```python
data, meta_data = ts.get_intraday('NSE:NOCIL',interval='1min',outputsize='full')

data.index=data.index.to_series().apply(lambda x:x[:-3])
```

```python
data_final=data_final.merge(data, how='inner', left_index=True, right_index=True)
```

```python
data_final_new=data_final.copy()
```

```python
data_final_new['bb_value']=0
```

```python
data_final_new
```

```python
for index, row in data_final_new.iterrows():
    if(row['4. close']>row['Real Middle Band']):
        data_final_new.at[index,'bb_value'] =(row['4. close']-row['Real Middle Band'])*100/(row['Real Upper Band']-row['Real Middle Band'])
    elif (row['4. close']<row['Real Middle Band']):
        data_final_new.at[index,'bb_value'] =(row['4. close']-row['Real Middle Band'])*100/(row['Real Middle Band']-row['Real Lower Band'])
    else:
        data_final_new.at[index,'bb_value'] =0
```

```python
data_final_new.to_csv(r'nocil.csv')
```
